```python
import ast

#ast have one method ast.literal_eval -> whos whork is to convert the
#string list to list of list 
#so here in the convert function its doing the same with the genres obj
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l    
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    l=[]
    counter = 0;
    for i in ast.literal_eval(obj):
        if counter != 3:
           l.append(i['name'])
           counter+=1
        else:
            break
    return l 
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l    
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
#now since we are going too convert all these rows into a single col of string
# so for not making so many keywords we have to merge the words of every col
#for eg: we are searching for [Sam Worthington] and we had not mergered these two grp 
#we are going to have two tags for 1 single person
#similary for the [Sam Mendes] 
#since sam is common in both of them it may also start to suggest the wrong mivie to the user
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] 
    for i in movies_list:
          print(new_df.iloc[i[0]].title)                               
```

```python
recommend('Avatar')
```
